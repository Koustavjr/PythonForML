```python
import pandas as pd
bios=pd.read_csv('./warmup-data/bios.csv')
bios.head()
bios_new=bios.copy()
```

```python
bios_new=bios.copy()
bios_new['first_name']=bios_new['name'].str.split(' ').str[0]
bios_new
```

```python
bios_new['born_datetime']=pd.to_datetime(bios_new['born_date'])
bios_new.info()
```

```python
bios_new['born_year']=bios_new['born_datetime'].dt.year
bios_new[['name','born_year']]
```

```python
bios_new.to_csv('./warmup-data/bios_new.csv',index=False)
```

```python
bios['height_category']=bios['height_cm'].apply(
    lambda x: 'Short' if x <165 else('Average' if x<185 else 'Tall')
)
bios.head()
```

```python
def category(row):
    if row['height_cm']<175 and row['weight_kg']<70:
        return 'Lightweight'
    else:
        return 'HeavyWeight'

bios['category']=bios.apply(category,axis=1)
```

# Aggregating Data

```python
bios['born_city'].value_counts()
```

```python
bios[bios['born_country']=='FRA']['born_region'].value_counts()
```

```python
# group sort and stuffs
coffee['Units Sold']=coffee['Units Sold'].interpolate()
coffee.groupby(['Coffee Type'])['Units Sold'].sum()
coffee.fillna(5.00)
coffee.groupby(['Coffee Type']).agg({'Units Sold':'sum','price':'mean'})
```

```python
#pivot
pivot=coffee.pivot(columns='Coffee Type',index='Day',values='revenue')
pivot
pivot.sum()
```

```python
bios['born_date']=pd.to_datetime(bios['born_date'])
bios.groupby(bios['born_date'].dt.year)['name'].count()
```

```python
bios['born_date']=pd.to_datetime(bios['born_date'])
bios.groupby(bios['born_date'].dt.year)['name'].count().reset_index().sort_values('name',ascending=False)
```

# Advanced Functionality

```python
# shift rank rolling cumsum
import pandas as pd
coffee=pd.read_csv('./warmup-data/coffee.csv')
coffee['prev units sold']=coffee['Units Sold'].shift(1)
coffee
```

```python
bios['height_rank']=bios['height_cm'].rank()
bios.sort_values(['height_rank'],ascending=True).sample(10)[['name','height_rank']]
```

```python
coffee['cumulative_sold']=coffee['Units Sold'].cumsum()
coffee
```

```python
latte=coffee[coffee['Coffee Type']=='Latte'].copy()
# cumulative sum 3 days
latte['3 day']=latte['Units Sold'].rolling(3).sum()
latte
```
