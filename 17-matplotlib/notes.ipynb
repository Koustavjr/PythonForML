# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
plt.plot([1,2,3],[1,2,3])
plt.title("Our first graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
#plt.show()
# change font
plt.title("Font",fontdict={"fontname":"Comic Sans MS"})
plt.show() 
```

```python
# For only interger values
plt.xticks([1,2,3])
plt.yticks([1,2,3])
plt.show()
```

```python
# add legend and color
plt.plot([1,2,3],[1,2,3],label="2x",color="red",linewidth="4",
         marker=".",markersize="20",markeredgecolor="blue",
         
         )
```

```python
# short hand 
plt.plot([1,2,3],[1,2,3],'r*--')
```

```python
x2=np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2,x2**2)
```

```python
# resize your graph
plt.figure(figsize=(5,5),dpi=500)

plt.plot([1,2,3],[1,2,3],color="red",label='x')
# save graph
plt.savefig("mygraph.png",dpi=300)
plt.show()
```
